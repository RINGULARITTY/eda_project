```python
import pandas as pd

ratings = pd.read_csv("./data/ratings.csv")
ratings.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

user_counts = ratings.groupby('user').count().reset_index()[["user", "mark"]]
user_counts.columns = ['user', 'num_purchases']

user_counts_sorted = user_counts.sort_values(by='num_purchases', ascending=False).head(100)

x = user_counts_sorted['user']
y = user_counts_sorted['num_purchases']

plt.figure(figsize=(22, 10), dpi=180)
sns.barplot(x=x, y=y)
plt.title('Distribution des quantités d\'achats par utilisateur')
plt.xlabel('Utilisateur')
plt.ylabel('Nombre d\'achats')
plt.show()
```

```python
frequency_counts = user_counts.groupby('num_purchases').count().reset_index()

frequency_counts.columns = ['num_purchases', 'num_users']

x = frequency_counts['num_purchases']
y = frequency_counts['num_users']

plt.figure(figsize=(22, 10), dpi=180)
sns.barplot(x=x, y=y)
plt.title('Fréquence d\'achats par nombre d\'utilisateurs')
plt.xlabel('Nombre d\'achats')
plt.ylabel('Nombre d\'utilisateurs')
plt.show()
```
