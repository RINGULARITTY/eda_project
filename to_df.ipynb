```python
import rapidjson as json
import pandas as pd

with open("./data/categories.json", "r") as f:
    categories = json.load(f)

df_categories = pd.DataFrame(list(categories.items()), columns=["category", "key"])
df_categories.to_csv("./data/categories.csv", index=False)
df_categories
```

```python
with open("./data/features.json", "r") as f:
    features = json.load(f)

df_features = pd.DataFrame(list(features.items()), columns=["feature", "key"])
df_features.to_csv("./data/features.csv", index=False)
df_features
```

```python
with open("./data/products.json", "r") as f:
    products = json.load(f)

df_products = pd.DataFrame(products)
df_products.to_csv("./data/products.csv", index=False)
df_products
```

```python
with open("./data/ratings.json", "r") as f:
    ratings = json.load(f)

df_ratings = pd.DataFrame(ratings)
df_ratings = df_ratings.drop(["vote", "image"], axis=1)
df_ratings.to_csv("./data/ratings.csv", index=False)
df_ratings
```
